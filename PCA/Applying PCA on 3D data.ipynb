### Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
np.random.seed(2343243) #assigning random seed value
```

### Link:https://numpy.org/doc/stable/reference/random/generated/numpy.random.multivariate_normal.html

```python
mean_vec1=np.array([0,0,0]) #assigning mean values for three features
cov_mat1=np.array([[1,0,0],[0,1,0],[0,0,1]]) #assigning covariance values for three features
class1=np.random.multivariate_normal(mean_vec1,cov_mat1,100)#generating 100 data points
class1.shape
```

```python
mean_vec2=np.array([1,1,1])#assigning mean values for three features
cov_mat2=np.array([[1,0,0],[0,1,0],[0,0,1]]) #assigning covariance values for three features
class2=np.random.multivariate_normal(mean_vec2,cov_mat2,100)#generating 100 data points
class2.shape
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D,proj3d

fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111,projection='3d')
ax.plot(class1[:,0],class1[:,1],class1[:,2],'o')
ax.plot(class2[:,0],class2[:,1],class2[:,2],'^')
plt.show()
```

```python
all_data=np.concatenate((class1,class2))
pca=PCA(n_components=2)
transformed_data=pca.fit_transform(all_data)
transformed_data.shape
```

```python
pca.components_ #two axes which are important
```

```python
plt.plot(transformed_data[0:100,0],transformed_data[0:100,1],"o")
plt.plot(transformed_data[100:200,0],transformed_data[100:200,1],"^")
plt.show()
```

```python
x_approx=pca.inverse_transform(transformed_data)
x_approx
```

```python
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111,projection='3d')
ax.plot(x_approx[:,0],x_approx[:,1],x_approx[:,2],'^')
plt.show()
```
