### Visualizing hourly pattern of engagements on posts for data before 2015
- We can observe that number of likes and number of reactions increased drastically over the years.
- Mostly before 2015 prominent engagements were only likes and reactions.

```python
before2015.groupby('hour')[reaction].sum()
```

```python
before2015.groupby('hour')[reaction].sum().plot()
```

- Video content: The number of likes and and number of reactions can be seen at peak in the year 2014.
- Photo content: The number of likes and number of reactions saw a steady increase from year 2014 to year 2015.
- Link and status content: The number of likes and number of reactions saw a steady increase from year 2014 to year 2015.

```python
before2015[before2015['status_type']==3].groupby('hour')[reaction].sum().plot(title="Trend of video content before 2015 ")
```

```python
before2015[before2015['status_type']==1].groupby('hour')[reaction].sum().plot(title="Trend of photo content before 2015 ")
```

```python
before2015[before2015['status_type']==0].groupby('hour')[reaction].sum().plot(title="Trend of link content before 2015 ")
```

```python
before2015[before2015['status_type']==2].groupby('hour')[reaction].sum().plot(title="Trend of status content before 2015 ")
```

### Visualizing yearwise growth of engagements on posts for data after 2015
- We can observe that number of comments increased drastically over the years.
- Unlike before 2015, other engagements also show an increasing trend over the years. 

```python
after2015.groupby('hour')[reaction].sum()
```

```python
after2015.groupby('year')[reaction].sum().plot()
```

- Video content: The video content has more number of comments and number of comments increases over the number of years.
- Photo content: The number of reactions and likes increases over the years and are comparatively more than other engagements.
- Status content: The number of reactions and likes increases over the years and are comparatively more than other engagements.
- Link content: The number of reactions and likes increases and then decreases and peak can be seen at year 2017 and are comparatively more than other engagements.

```python
after2015[after2015['status_type']==3].groupby('year')[reaction].sum().plot(title="Trend of video content after 2015 ")
```

```python
after2015[after2015['status_type']==1].groupby('year')[reaction].sum().plot(title="Trend of photo content after 2015 ")
```

```python
after2015[after2015['status_type']==2].groupby('year')[reaction].sum().plot(title="Trend of status content after 2015 ")
```

```python
after2015[after2015['status_type']==0].groupby('year')[reaction].sum().plot(title="Trend of link content after 2015 ")
```
